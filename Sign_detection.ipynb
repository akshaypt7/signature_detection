```python
import cv2 as cv
import numpy as np
import imutils
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python

img_path = '/content/drive/MyDrive/Curiosity/23608978.JPG'
img = cv.imread(img_path)

# cv.imshow('window',img)


# resize
resized_img= cv.resize(img, (500,500), interpolation=cv.INTER_CUBIC)
cv2_imshow(resized_img)
```

```python
gray_img = cv.cvtColor(resized_img.copy(),cv.COLOR_BGR2GRAY)
# cv2_imshow(gray_img)
```

```python
canny = cv.Canny(resized_img.copy(),150,255)
cv2_imshow(canny)
```

```python
# thershold, thresh = cv.threshold(gray_img.copy(),120, 255, cv.THRESH_BINARY)

# cv2_imshow(thresh)
```

```python
contours, hierarchies = cv.findContours(canny, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
```

```python
image = cv.drawContours(resized_img.copy(), contours, -1, (0, 255, 0), 2)
cv2_imshow(image)
```

```python
# extracting the biggest contour
biggest_contour = max(contours, key = cv.contourArea)

sign_img = cv.drawContours(resized_img.copy(),biggest_contour,-1,(0,255,0),3)
cv2_imshow(sign_img)
```

```python
x, y, w, h = cv.boundingRect(biggest_contour)
x, y, w, h
```

```python
# Drawing a rectangle over the biggest contour

x, y, w, h = cv.boundingRect(biggest_contour)

box_sign_img = cv.rectangle(resized_img.copy(),(x-w-30,y-w-20),(x+w+50,y+h+30),(0,255,0),3)
cv2_imshow(box_sign_img)
```
